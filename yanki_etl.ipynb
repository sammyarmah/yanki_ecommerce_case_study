DATA CLEANING AND TRANSFORMATION

```python
### drop missing values
yanki_df.dropna(subset = ['Order_ID', 'Customer_ID'], inplace=True)
```

```python
yanki_df.info()
```

```python
## convert order_date from string to datetime object
yanki_df['Order_Date'] = pd.to_datetime(yanki_df['Order_Date'])
```

```python
### creating the customer table
customer = yanki_df[['Customer_ID', 'Customer_Name', 'Email', 'Phone_Number']].copy().drop_duplicates().reset_index(drop=True)
customer.head()
```

```python
yanki_df.columns
```

```python
product = yanki_df[['Product_ID', 'Product_Name', 'Brand', 'Category', 'Price']].copy().drop_duplicates().reset_index(drop=True)
product.head()
```

```python
shipping_address = yanki_df[['Customer_ID', 'Shipping_Address', 'City', 'State', 'Country','Postal_Code']].copy().drop_duplicates().reset_index(drop=True)
shipping_address.head()

shipping_address.index.name = 'shipping_ID'
shipping_address = shipping_address.reset_index()

shipping_address.head()
```

```python
order = yanki_df[['Order_ID','Customer_ID', 'Product_ID', 'Quantity', 'Total_Price', 'Order_Date']].copy().drop_duplicates().reset_index(drop=True)
order.head()
```

```python
payment_method = yanki_df[['Order_ID', 'Payment_Method', 'Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)
payment_method.head()
```

```python
# save the table to csv
customer.to_csv(r'dataset\cleaned_data\customer.csv', index=False)
product.to_csv(r'dataset\cleaned_data\product.csv', index=False)
order.to_csv(r'dataset\cleaned_data\order.csv', index=False)
shipping_address.to_csv(r'dataset\cleaned_data\shipping_address.csv', index=False)
payment_method.to_csv(r'dataset\cleaned_data\payment_method.csv', index=False)
```
